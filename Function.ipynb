### Для понимания того с какой функцией мы имеем дело построим график:

```python
import matplotlib.pyplot as plt
import math
import sympy
from sympy import solveset
from sympy import cos, sin
```

```python
list_x = []
list_y =[]
x = - 50
y = - 50
while x <= 50:
    y = (-12 * (x**4) * math.sin(math.cos(x)) - 18*(x**3) + 5*(x**2) + (10*x) - 30)
    list_y.append(y)
    list_x.append(x)
    x += 0.1
    y += 0.1


plt.plot( list_x, list_y, 'b-')
plt.grid()
plt.show()
```

Здорово! Это чёртова тригонометрическая функция, да еще и увеличивающая свою амплетуду в зависимости от удаления от 0.

Ладно, давайте рассмотрим эту функцию по частям:

```python
list_x = []
list_y =[]
x = - 50
y = - 50
while x <= 50:
    y = (-12 * (x**4) * math.sin(math.cos(x)) - 18*(x**3) + 5*(x**2) + (10*x) - 30)
    list_y.append(y)
    list_x.append(x)
    x += 0.1
    y += 0.1
    
plt.subplot(2, 2, 1)
y = [list_y[i] for i in range(0, 300)]
x = [list_x[i] for i in range(0, 300)] 
plt.plot(x, y, 'b-')
plt.title("x ∈ [-50, -20]")
plt.grid()

plt.subplot(2, 2, 2)
y = [list_y[i] for i in range(700, 1000)]
x = [list_x[i] for i in range(700, 1000)] 
plt.plot(x, y, 'b-')
plt.title("x ∈ [20, 50]")
plt.grid()

plt.subplot(2, 2, 3)
y = [list_y[i] for i in range(400, 600)]
x = [list_x[i] for i in range(400, 600)] 
plt.plot(x, y, 'b-')
plt.title("x ∈ [-10, 10]")
plt.grid()

plt.subplot(2, 2, 4)
y = [list_y[i] for i in range(490, 511)]
x = [list_x[i] for i in range(490, 511)] 
plt.plot(x, y, 'b-')
plt.title("x ∈ [-1, 1]")
plt.grid()

plt.show()
```

Так-с, ну понятно что ничего непонятно... хе - хе

Ладно, если по серьезному. Мы видим, что точек экстремума бесконечное количество.
Функция не переодична, т.к. f(x)≠f(x+T), где Т - период функции.
Чтобы найти экстремумы функции продифференцируем ее, и в точках где f′(x) = 0, будут точки экстремума.

```python
x = sympy.Symbol('x')
expr = 12*(x**4)*sin(x)*cos(cos(x))-48*(x**3)*sin(cos(x))-54*(x**2)+10*x+10
diff = sympy.simplify(expr)
print(f'f′(x) = {diff}')
```

Получили производную. Здорово! Построим её график:

```python
list_x = []
list_y =[]
z_list = []
x = - 50
y = - 50

while x <= 50:
    x1 = (-12 * (x**4) * math.sin(math.cos(x)) - 18*(x**3) + 5*(x**2) + (10*x) - 30)
    list_x.append(x1)
    list_y.append(y)
    z = 12*(x**4)*math.sin(x)*math.cos(math.cos(x))-48*(x**3)*math.sin(math.cos(x))-54*(x**2)+10*x+10
    z_list.append(z)
    x += 0.1
    y += 0.1

plt.subplot(2, 2, 1)
plt.plot( list_y, list_x, 'b-')
plt.title('f(x)')
plt.grid()


plt.subplot(2, 2, 2)
plt.plot( list_y, z_list, 'r-')
plt.title('Производная')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot( list_y, list_x, 'b-', list_y, z_list, 'r-')
plt.grid()
plt.show()
```

Попробуем найти "нули" производной:

```python
diff = 12*(x**4)*sin(x)*cos(cos(x))-48*(x**3)*sin(cos(x))-54*(x**2)+10*x+10
x = sympy.symbols('x')
solveset(diff, x)
```

Ответ довольно понятен, тк число экстремумов бесконечное число.

Попробуем исследовать функцию на отрезке x ∈ [-1, 1]
График: 

```python
list_x = []
list_y =[]
z_list = []
x = - 1
y = - 1

while x <= 1:
    x1 = (-12 * (x**4) * math.sin(math.cos(x)) - 18*(x**3) + 5*(x**2) + (10*x) - 30)
    list_x.append(x1)
    list_y.append(y)
    z = 12*(x**4)*math.sin(x)*math.cos(math.cos(x))-48*(x**3)*math.sin(math.cos(x))-54*(x**2)+10*x+10
    z_list.append(z)
    x += 0.1
    y += 0.1
plt.plot( list_y, list_x, 'b-', list_y, z_list, 'r-')
plt.text(-1, 6, "Производная", fontsize=15)
plt.text(0.45, -34, "f(x)", fontsize=15)
plt.grid()
plt.show()
```

Видим, что производная пересекает ось X в 2х местах --> у нас 2 точки экстремума.
Найдем их методом "тыка":

```python
x = -1
points_extremum = []
while x <= 1:
    y = 12*(x**4)*math.sin(x)*math.cos(math.cos(x))-48*(x**3)*math.sin(math.cos(x))-54*(x**2)+10*x+10
    if y <= 0.0004 and y >= -0.0004:
        point = (x, y)
        points_extremum.append(point)
    x += 0.00001
print(points_extremum)
```

Мы получили две пары точек, между которыми лежат нули этой производной. Немного округлив получим 2 точки экстремума:
При x = -0.392675 и при x = 0.454575. Найдем f(x) при этих значениях:

```python
x_points = [-0.392675, 0.454575]
extremums = []
for i in range(len (x_points)):
    x = x_points[i]
    y = 12*(x**4)*math.sin(x)*math.cos(math.cos(x))-48*(x**3)*math.sin(math.cos(x))-54*(x**2)+10*x+10
    extremums.append((x, y))
print(extremums)
```
